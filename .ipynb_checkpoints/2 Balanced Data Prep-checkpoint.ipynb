```python
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import RandomOverSampler

# Load the cleaned and normalized dataset
df = pd.read_csv("Dataset/2 Cleaned_Predictive_Maintenance.csv")

# Bar charts for categorical features
for col in ['Type', 'Machine failure']:
    df[col].value_counts().plot(kind='bar', title=f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python

# Separate features (X) and target (y)
# Make sure to include only the features you want to use.
X = df.drop(columns=['Machine failure'])
y = df['Machine failure']

# Initialize RandomOverSampler
ros = RandomOverSampler(random_state=42)

# Apply Random Over-Sampling to balance the classes
X_resampled, y_resampled = ros.fit_resample(X, y)

# Combine the resampled features and target into a single DataFrame
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), 
                          pd.DataFrame(y_resampled, columns=['Machine failure'])], axis=1)

# Save the balanced dataset
df_resampled.to_csv("Dataset/3 Balanced_Predictive_Maintenance.csv", index=False)

print("Random over-sampling complete. The balanced dataset is saved as 'Balanced_Predictive_Maintenance.csv'.")
```
